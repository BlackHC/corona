```python
data = pd.read_excel("./data/census-2011-population-5year-age-ukdistricts.xls", sheet_name=1, index_col=[0,1,2])
data
```

```python
london_data = data.xs("London", level="Area")
london_data
```

```python
london_age_buckets = london_data.reset_index().transpose()[0]["Age 0 to 4":]
london_age_buckets
```

```python
london_age_buckets_by_decade = {}
for i in range(9):
    key = f"{i*10} <= age < {i*10+10}"
    london_age_buckets_by_decade[key] = london_age_buckets[2*i] + london_age_buckets[2*i+1]
london_age_buckets_by_decade["age >=90"] = london_age_buckets[-1]

london_age_buckets_by_decade
```
